```python

# Проверка распределения выбранных числовых колонок в DataFrame
def analyze_distribution(df, target_col=None, top_n=10):
    # Основная информация о числовых колонках
    print("Основные статистические характеристики:")
    print(df.describe())


    # Выбор только числовых колонок
    num_cols = df.select_dtypes(include=['number']).columns

    # Если указана целевая переменная, ищем корреляцию
    if target_col and target_col in df.columns:
        correlations = df[num_cols].corrwith(df[target_col]).abs().sort_values(ascending=False)
        print("Топ коррелирующих колонок с целевой переменной:")
        print(correlations.head(top_n))
        num_cols = correlations.head(top_n).index
    else:
        # Если целевая переменная не указана, берем топ по стандартному отклонению
        stds = df[num_cols].std().sort_values(ascending=False)
        print("Топ колонок с наибольшей вариативностью:")
        print(stds.head(top_n))
        num_cols = stds.head(top_n).index

    # Построение гистограмм и боксплотов
    cols_per_plot = 4
    for i in range(0, len(num_cols), cols_per_plot):
        subset_cols = num_cols[i:i + cols_per_plot]

        # Гистограммы
        plt.figure(figsize=(16, 10))
        for j, col in enumerate(subset_cols):
            plt.subplot(2, 2, j + 1)
            sns.histplot(df[col], kde=True, bins=30, color='blue', edgecolor='black')
            plt.title(f'Распределение {col}')
            plt.xlabel(col)
            plt.ylabel('Частота')
        plt.tight_layout()
        plt.show()

        # Боксплоты
        plt.figure(figsize=(16, 10))
        for j, col in enumerate(subset_cols):
            plt.subplot(2, 2, j + 1)
            sns.boxplot(x=df[col], color='lightblue')
            plt.title(f'Боксплот для {col}')
            plt.xlabel(col)
        plt.tight_layout()
        plt.show()
analyze_distribution(df, target_col="target", top_n=10)

```

```python
# удаляем одинаковые колонки

def del_dup(df):
  unique_columns = []
  for column in df.columns:
    if not any(df[column].equals(df[col]) for col in unique_columns):
      unique_columns.append(column)
  return df[unique_columns]

df = del_dup(df)
df.head(10)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

def encode_categorical_columns(df):
    # Создаем экземпляр LabelEncoder
    encoder = LabelEncoder()
    
    # Применяем кодирование к каждому категориальному столбцу
    for col in df.select_dtypes(include=['object']).columns:
        # Пропускаем NaN значения и кодируем только уникальные значения
        df[col] = df[col].fillna('missing')  # Заменяем NaN на 'missing' перед кодированием
        df[col] = encoder.fit_transform(df[col])  # Применяем LabelEncoder
        
    return df


df1 = encode_categorical_columns(df)
print(df1)
```

```python
from sklearn.ensemble import IsolationForest

def remove_outliers_isolation_forest(df1, contamination=0.1):
    model = IsolationForest(contamination=contamination, random_state=42)
    
    # Применяем только к числовым колонкам
    num_data = df1.select_dtypes(include=np.number)
    model.fit(num_data)
    
    # Предсказываем "нормальность" данных (-1: выброс, 1: нормально)
    df1['anomaly'] = model.predict(num_data)
    df1 = df1[df1['anomaly'] == 1].drop(columns=['anomaly'])
    return df1

df2 = remove_outliers_isolation_forest(df1, contamination=0.1)  # Удаляем 5% самых аномальных строк
print(f"Размер датафрейма после обработки: {df1.shape}")
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Отбор признаков на основе корреляции с целевой переменной
def select_features_by_correlation(df3, target, threshold=0.1):
    correlation = df.corr()[target].abs()
    selected_features = correlation[correlation > threshold].index.tolist()
    selected_features.remove(target)
    return selected_features

# Определяем категориальные признаки
cat_features = df3.select_dtypes(include=['object']).columns.tolist()
# Определяем категориальные признаки
cat_features = df3.select_dtypes(include=['object']).columns.tolist()

if not cat_features:
    print("Нет категориальных признаков для обработки.")
else:
    imputer = SimpleImputer(strategy='most_frequent')
    df3[cat_features] = imputer.fit_transform(df3[cat_features])

    for col in cat_features:
        le = LabelEncoder()
        df3[col] = le.fit_transform(df3[col])

# Обработка пропусков в числовых данных
num_features = df3.select_dtypes(include=['float64', 'int64']).columns.tolist()
df3[num_features] = SimpleImputer(strategy='mean').fit_transform(df3[num_features])



# Отбор признаков на основе корреляции
selected_features = select_features_by_correlation(df3, target='target', threshold=0.1)

# Разделение данных на признаки и целевую переменную
X = df3[selected_features]
y = df3['target']

# Балансировка классов с помощью SMOTE
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X_balanced, y_balanced,
    test_size=0.2,
    random_state=42
)

# Создание и настройка модели XGBoost
model = xgb.XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    scale_pos_weight=len(y_train[y_train == 0]) / len(y_train[y_train == 1]), # Учет дисбаланса классов
    eval_metric='auc',
)

# Обучение модели
model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=10
)

# Получение прогнозов вероятностей
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Оценка качества модели
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f'ROC-AUC на тестовом датасете: {roc_auc}')

# Сравнение предсказаний с реальными значениями
predictions = pd.DataFrame({
    'Actual': y_test,
    'Predicted_Probability': y_pred_proba
})
predictions['Predicted_Class'] = (predictions['Predicted_Probability'] >= 0.5).astype(int)

# Печать классификационного отчета
print("\nClassification Report:")
print(classification_report(y_test, predictions['Predicted_Class']))

# Вывод первых строк для анализа
print("\nСравнение прогнозов и реальных значений:")
print(predictions.head())

# Построение boxplot для наиболее важных признаков
feature_importances = model.feature_importances_
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

# Отбираем топ-N важных признаков
top_features = importance_df.head(10)['Feature'].tolist()

# Построение boxplot для этих признаков
plt.figure(figsize=(15, 10))
for i, feature in enumerate(top_features):
    plt.subplot(2, 5, i + 1)
    df3.boxplot(column=feature, by='target', grid=False)
    plt.title(feature)
    plt.suptitle('')  # Убираем общий заголовок
    plt.xlabel('Target')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
import optuna
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
import pandas as pd
import numpy as np



# Определяем категориальные признаки
cat_features = df3.select_dtypes(include=['object']).columns.tolist()

if not cat_features:
    print("Нет категориальных признаков для обработки.")
else:
    imputer = SimpleImputer(strategy='most_frequent')
    df3[cat_features] = imputer.fit_transform(df3[cat_features])

    for col in cat_features:
        le = LabelEncoder()
        df3[col] = le.fit_transform(df3[col])

# Обработка пропусков в числовых данных
num_features = df3.select_dtypes(include=['float64', 'int64']).columns.tolist()
df3[num_features] = SimpleImputer(strategy='mean').fit_transform(df3[num_features])

# Разделение данных на признаки и целевую переменную
target_column = 'target'  # Замените на имя вашей целевой переменной
X = df3.drop(columns=[target_column])
y = df3[target_column]

# Балансировка классов с помощью SMOTE
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X_balanced, y_balanced,
    test_size=0.2,
    random_state=42
)

# Оптимизационная функция
def objective(trial):
    # Гиперпараметры
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 300),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'scale_pos_weight': len(y_train[y_train == 0]) / len(y_train[y_train == 1])
    }
    
    # Создание модели
    model = xgb.XGBClassifier(eval_metric='auc', **params)
    model.fit(X_train, y_train)
    
    # Предсказания
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    # ROC-AUC
    return roc_auc_score(y_test, y_pred_proba)

# Запуск оптимизации
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)

# Лучшие параметры
print("Лучшие параметры:", study.best_params)
print("Лучший ROC-AUC:", study.best_value)

# Обучение финальной модели на лучших параметрах
best_params = study.best_params
final_model = xgb.XGBClassifier(eval_metric='auc', use_label_encoder=False, **best_params)
final_model.fit(X_train, y_train)

# Оценка на тестовой выборке
y_pred_proba = final_model.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_pred_proba)

print(f'\nROC-AUC на тестовой выборке: {roc_auc}')
print("\nClassification Report:")
print(classification_report(y_test, (y_pred_proba >= 0.5).astype(int)))

# Сравнение предсказаний и реальных значений
predictions = pd.DataFrame({
    'Actual': y_test,
    'Predicted_Probability': y_pred_proba,
    'Predicted_Class': (y_pred_proba >= 0.5).astype(int)
})
print("\nПрогнозы:")
print(predictions.head())
```
